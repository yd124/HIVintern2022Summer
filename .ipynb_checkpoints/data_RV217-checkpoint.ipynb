```python
n = len(patients)
nrows = math.ceil(n/3)

fig, axs = plt.subplots(ncols=3,nrows=nrows,figsize=(16,20),tight_layout=True)

for k in range(n):
    ftr = patients[k]    
    
    i = k // 3
    j = k% 3
    
    subdf = df.groupby(['ID']).get_group(ftr)
    x = list(subdf.days)
    axs[i][j].scatter(x,subdf['log10VL'],label=ftr)
    axs[i][j].legend(loc='best')

plt.show()
```
