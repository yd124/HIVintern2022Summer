* T: target cells （CD4$^+$ T cells)
* I: productively infected CD4$^+$ T cells,
* V: free virus

\begin{align}
\frac{dT}{dt} &= \lambda -dT -\beta(t)TV, \quad T(0) =T_0\\
\frac{dI}{dt} &= \beta(t) TV - \delta I,\quad  I(0)= I_0 \\
\frac{dV}{dt} &= pI - cV, \quad V(0) = V_0
\end{align}
where 
\begin{align}
\beta(t) = 
\begin{cases}
\beta_0, \quad t\leq \tau \\
\beta_{\infty} + (\beta_0 - \beta_{\infty}) e^{-k(t-\tau)}, \quad t>\tau
\end{cases}
\end{align}

Determine the parameters $(\beta_0,\beta_{\infty},\lambda, d, \delta, p, c, k, \tau)$ so that it can fit the real data


```python
# b = beta(t,tau,k,beta_0,beta_ifty)

def fT(lda,d,T,b,V):
    return lda - d * T - b*T*V

def fI(b,T,V,dlt,I):
    return b*T*V-dlt*I

def fV(p,I,c,V):
    return p*I-c*V
```

```python
def beta(t, tau, k,beta_0,beta_ifty):
    if t <= tau:
        return beta_0
    else:
        return beta_ifty + (beta_0-beta_ifty)*np.exp(-k*(t-tau))
```

```python
def runge_kutta(T,V,I, h,b,lda,d,dlt,p,c):
#     h,b,lda,d,dlt,p,c = param

    k1 = h*fT(lda,d,T,b,V)
    k2 = h*fT(lda,d,T+k1/2,b,V)
    k3 = h*fT(lda,d,T+k2/2,b,V)
    k4 = h*fT(lda,d,T+k3,b,V)
    T_new = T+k1/6+k2/3+k3/3+k4/6
    
    k1 = h*fI(b,T,V,dlt,I)
    k2 = h*fI(b,T,V,dlt,I+k1/2)
    k3 = h*fI(b,T,V,dlt,I+k2/2)
    k4 = h*fI(b,T,V,dlt,I+k3)
    I_new = I+k1/6+k2/3+k3/3+k4/6
    
    k1 = fV(p,I,c,V)
    k2 = fV(p,I,c,V+k1/2)
    k3 = fV(p,I,c,V+k2/2)
    k4 =fV(p,I,c,V+k3)
    V_new = V+k1/6+k2/3+k3/3+k4/6
    
    return [max(T_new,0),max(I_new,0),max(V_new,0)]
    
    
```
