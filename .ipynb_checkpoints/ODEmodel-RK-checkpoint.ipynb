```python
# fT = dT/dt
def fT(lda,d,T,b,V):
    return lda - d * T - b*T*V

# fI = dI/dt
def fI(b,T,V,dlt,I):
    return b*T*V-dlt*I

# fV = dV/dt
def fV(p,I,c,V):
    return p*I-c*V

def beta(t, tau, k,beta_0,beta_ifty):
    if t <= tau:
        return beta_0
    else:
        return beta_ifty + (beta_0-beta_ifty)*np.exp(-k*(t-tau))
```

```python
def runge_kutta(T,V,I, h,b,lda,d,dlt,p,c):

    T_k1 = h*fT(lda,d,T,b,V)
    T_k2 = h*fT(lda,d,T+T_k1/2,b,V)
    T_k3 = h*fT(lda,d,T+T_k2/2,b,V)
    T_k4 = h*fT(lda,d,T+T_k3,b,V)
    T_new = T+T_k1/6+T_k2/3+T_k3/3+T_k4/6
    
    I_k1 = h*fI(b,T,V,dlt,I)
    I_k2 = h*fI(b,T,V,dlt,I+I_k1/2)
    I_k3 = h*fI(b,T,V,dlt,I+I_k2/2)
    I_k4 = h*fI(b,T,V,dlt,I+I_k3)
    I_new = I+I_k1/6+I_k2/3+I_k3/3+I_k4/6
    
    V_k1 = fV(p,I,c,V)
    V_k2 = fV(p,I,c,V+V_k1/2)
    V_k3 = fV(p,I,c,V+V_k2/2)
    V_k4 =fV(p,I,c,V+V_k3)
    V_new = V+V_k1/6+V_k2/3+V_k3/3+V_k4/6
    
    return [max(T_new,0),max(I_new,0),max(V_new,0)]
    
    
```

```python
def predict_v(beta_0,beta_ifty, k, dlt, p, d, tau):
    # initial value 
    T_0 = 10**4  # ml
    I_0 = 0
    V_0 = 10**-3
    c = 23 # 1/day  p.2
    lda = d * T_0 
    
    t_end = 20
    h = 0.01
    N = int(t_end/h) + 1

    
    T = N * [T_0]
    I = N * [I_0]
    V = N * [V_0]
    
    i = 0
    t = h
    while t < t_end:
        b = beta(t,tau,k,beta_0,beta_ifty)
        T[i+1],I[i+1],V[i+1] = runge_kutta(T[i],I[i],V[i],h,b,lda,d,dlt,p,c)

        i += 1
        t += h
        
    return V
```
